# Question 1 


1.a. A descriptive analysis of the additives (columns named as “a” to “i”), which must include summaries of findings (parametric/non-parametric). Correlation and ANOVA, if applicable, is a must.

1.b. A graphical analysis of the additives, including a distribution study.


```python
# Cell 1

import pandas as pd
from pandas_profiling import ProfileReport
import scipy.stats as stats
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, HoverTool, RangeTool, Range1d
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Data exploration
df_ingredient = pd.read_csv('ingredient.csv')
profile = ProfileReport(df_ingredient)
profile.to_file(output_file="profile_ingredient.html")

display(df_ingredient.describe())
```

```python
# Cell 2

# Normality test using Shapiro–Wilk test
for column in df_ingredient.columns:
    values = df_ingredient[column]
    
    # Perform the Shapiro-Wilk test for normality
    _, p_value = stats.shapiro(values)
    
    # Determine if the values are parametric or non-parametric based on p-value
    if p_value > 0.05:
        print(f"{column}: {p_value} Parametric")
    else:
        print(f"{column}: {p_value} Non-parametric")
```

1.c. A clustering test of your choice (unsupervised learning), to determine the distinctive number of formulations present in the dataset.

```python
# Cell 3

# K-means clustering implementation
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from bokeh.layouts import column
from bokeh.models import ColumnDataSource, HoverTool, RangeTool, Range1d
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

import plotly.express as px
import seaborn as sns
# %pylab inline

output_notebook()

df_kmeans_cluster = df_ingredient

# scale data
scalar = MinMaxScaler()
scalar.fit(df_ingredient)
df_kmeans = scalar.transform(df_ingredient)

# Loop for 50 K values.
sse = []
k_rng = range(1,50)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df_kmeans)
    sse.append(km.inertia_)

# Save inertia and SSE results as dataframe
df1 = pd.DataFrame(k_rng, columns=['K']).reset_index()
df2 = pd.DataFrame(sse, columns=['SSE']).reset_index()
df1 = pd.merge(df1, df2, right_on = 'index', left_on = 'index').set_index('index')

# Plot elbow method figure. Line chart.
p = figure(plot_width=800, plot_height=400)
source = ColumnDataSource(df1)
p.line(x='K', y='SSE', source=source, line_width=2, line_color='blue')
p.add_tools(HoverTool(tooltips=[("K", "@K"),("Sum of Square Error", "@SSE")]))
p.title.text = 'Elbow Method'
p.title.text_font_size = '15pt'

show(column(p))

# K=25 selected and the custering resutls below
km = KMeans(n_clusters=25)
km.fit(df_kmeans)
y_predicted = km.fit_predict(df_kmeans)

# Save cluster result back to raw data
df = pd.DataFrame(y_predicted, columns=['cluster'])
df_kmeans_cluster = pd.concat([df_kmeans_cluster,df], axis='columns')
# df_kmeans_cluster.to_csv('ingredient_cluster_25.csv')

# Cluster centroid
# display(km.cluster_centers_)
```

# Question 3
3.a. What is the probability of the word “data” occurring in each line ?

3.b. What is the distribution of distinct word counts across all the lines ?


```python
# Cell 6

import pytesseract
from PIL import Image
import re

# Get paragraph in lines as provided in PDF
text = pytesseract.image_to_string(Image.open('Q3_paragraph.jpg'))

paragraph = text.splitlines()

paragraph_cleaned = []

# Remove unwanted symbols, retain proper words and comvert to lower case letters
for line in paragraph:
    paragraph_cleaned.append(re.sub(r'[^\w\s]', '', line.lower()))

column_names = ['Line','Total word','Count of data', 'Pos. of data', 'Count of analytics', 'Pos. of analytics','Unique words']
df_paragraph_results = pd.DataFrame(columns = column_names)
df_paragraph_results

for line, sentence in enumerate(paragraph_cleaned):
    total_word_line = len(sentence.split())
    data_pos = []
    analytics_pos = []
    Unique_words = []
    for i,word in enumerate(sentence.split()):
        if (word == 'data'): 
            data_pos.append(i)
        if (word == 'analytics'):
            analytics_pos.append(i) 
        if word not in Unique_words:
            Unique_words.append(word)
    df_line_result = pd.DataFrame([[line, total_word_line, len(data_pos), data_pos, len(analytics_pos), analytics_pos, len(Unique_words)]], columns=column_names)
    df_paragraph_results = pd.concat((df_paragraph_results, df_line_result), ignore_index = True, axis='rows')
    
df_paragraph_results
```

```python
# Cell 7
# Answering Q3.a.

df_paragraph_results['Q3A_Probability'] = df_paragraph_results['Count of data']/df_paragraph_results['Total word']
print('The probability of obtaining the word data in each line: ',round(100*df_paragraph_results['Q3A_Probability'].mean(),2), ' %')
```

```python
# Cell 8
# Answering Q3.b.

total = df_paragraph_results['Unique words'].sum()
average = df_paragraph_results['Unique words'].mean()
print('Total distinct word counts across all the lines: ', total)
print('Average distinct words per lines: ', round(average,2))
```

3.c. What is the probability of the word “analytics” occurring after the word “data” ?

```python
# Cell 9
# Answering Q3.c.
# Check for occurance of word analytics after data, using position of word 'data' and 'analytics' in each line.

Total_Analytics_after_Data = 0

for i, val in enumerate(df_paragraph_results['Pos. of analytics']):
    A_after_D =[]
    if len(val)>0:
        for x in val:
#           If analytics does not occur at beginning of line, then check if there was 'data' before word 'analytics'.
            if x!=0:
                if x-1 in df_paragraph_results['Pos. of data'][i]:
                    A_after_D.append('True')
            else:
#           If analystics occurs at beginning of line, check if there was 'data' at end of previous line.
                if df_paragraph_results['Total word'][i-1]-1 in df_paragraph_results['Pos. of data'][i-1]:
                    A_after_D.append('True')
                
    Total_Analytics_after_Data = Total_Analytics_after_Data +len(A_after_D)
    
print("Total time 'analytics' appeared right after word 'data': ",Total_Analytics_after_Data)

Total_word = df_paragraph_results['Total word'].sum()
Q3c_A = 100*df_paragraph_results['Count of data'].sum() / Total_word
Q3c_B = 100*Total_Analytics_after_Data / Total_word
Q3c_Final = 100 * Q3c_B / Q3c_A
print("Probability of word 'data': ", round(Q3c_A,2), ' %')
print("Probability of 'data' followed by 'analytics': ", round(Q3c_B,2), ' %')
print("Probability of the word 'analytics' occurring after the word 'data': ", round(Q3c_Final,2), ' %')
```
